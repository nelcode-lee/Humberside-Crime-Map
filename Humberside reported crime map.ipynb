```python
import plotly.express as px 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('2024-01-humberside-street.csv')
```

```python
df.head()
```

```python
fig = px.scatter_mapbox(df, lon=df['Longitude'],lat=df['Latitude'],color=df['Crime type'], zoom = 3, width =1200, height = 900, title = 'Humberside - Reported crime by location')
fig.update_layout(mapbox_style='open-street-map')
fig.show()
```
